```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'E:/udemy/Data_kmeans/bank.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
mean_v1 = np.mean(df['V1'])
mean_v2 = np.mean(df['V2'])
print(mean_v1,mean_v2)
```

```python
std_v1 = np.std(df['V1'])
std_v2 = np.std(df['V2'])
print(std_v1,std_v2)
```

```python
df.describe()
```

```python
des=pd.DataFrame(df.describe())
des.to_csv('raw_data.csv') 
```

```python
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(df['V1'],df['V2'])
plt.scatter(mean_v1,mean_v2,s=1000,alpha=0.5)
plt.scatter(std_v1,std_v2,s=1000,alpha=0.5)
plt.text(mean_v1,mean_v2,'Mean')
plt.text(std_v1,std_v2, 'Std')
plt.savefig('fig2')
```

```python
from sklearn.cluster import KMeans

#print(GDP_happy.shape)

km_res=KMeans(n_clusters=2).fit(df)
clust=km_res.cluster_centers_
plt.scatter(df['V1'],df['V2'],c=km_res.labels_.astype(float))
plt.scatter(clust[:,0],clust[:,1],s=1000,alpha=0.5)
plt.scatter(mean_v1,mean_v2,s=100,color='white',alpha=0.7)
plt.scatter(std_v1,std_v2,s=100,color='white',alpha=0.7)
plt.text(mean_v1,mean_v2,'Mean')
plt.text(std_v1,std_v2, 'Std')
plt.savefig('fig3')
```
